## break, else en continue

Soms wil je wat meer vrijheid in de manier waarop je Python code door een loop heen beweegt. De __for__ loop hebben we al veel toegepast, ook hierboven. Als je niet ingrijpt, wordt de code in je __for__ loop steeds helemaal uitgevoerd. Maar dat is misschien niet altijd precies wat je wilt. Alleen al voor de tijdswinst kan het soms handig zijn de berekeningen voortijdig af te breken.

```python
# Priemgetallen zoeken.

for n in range(2, 100):
    for x in range(2, n):
        if n % x == 0:
            break
    else:
        print(n)
        
```

In het [priemgetallen](https://nl.wikipedia.org/wiki/Priemgetal) voorbeeld hierboven zie je een paar speciale dingen gebeuren. We hebben twee __for__ loops in elkaar. Dat noemen we __nesten__, als je twee loops in elkaar hebt lopen. De __n__ loop is de buitenste loop, en daarbinnen is de __x__ loop genest. Het __%__ symbooltje noemen we ook wel de __modulus__ operator. Deze geeft aan wat de rest is na een deling. Als er geen rest is, dus als het resultaat van n % x gelijk is aan 0, dan zijn de getallen deelbaar door elkaar en is n dus duidelijk geen [priemgetal](https://nl.wikipedia.org/wiki/Priemgetal). Het __break__ commando zegt dan dat de binnenste __for__ loop mag stoppen. Het heeft geen zin om door te rekenen, het is toch geen priemgetal. De __else__ instructie daarna is een hele speciale. Die wordt uitgevoerd als de __for__ loop helemaal klaar is met door alle getallen te lopen, en ook alleen maar dan. 

Zie je ook hoe kort en overzichtelijk de code is? Dat is wat Python ook zo mooi en krachtig maakt. Je hoeft maar een 0 toe te voegen en de code rekent de priemgetallen tot 1000 voor je uit. Probeer het maar eens!

En stel nu eens dat we onze [priemgetallen](https://nl.wikipedia.org/wiki/Priemgetal) zelf ook eerst in een lijst zouden willen parkeren. Dan kunnen we er later ook nog makkelijker bij, toch? De code zou er dan als volgt uit kunnen zien:

```python
# Priemgetallen zoeken.

priemgetallen = []

for n in range(2, 100):
    for x in range(2, n):
        if n % x == 0:
            break
    else:
        priemgetallen.append(n)
        
print("We hebben %d priemgetallen gevonden." 
      % len(priemgetallen))

# Hoeveel willen we er zien?
n = 5

print("Dit zijn de eerste", n, ":")

for idx in range(0, n):
    print(priemgetallen[idx])
```

Hierboven maken we dus eerst een lege lijst (__priemgetallen = []__) en daarna plakken we er steeds weer eentje achter als we een nieuw priemgetal hebben ontdekt (__priemgetallen.append(n)__).

Met de __len__ functie kunnen we te allen tijde zien hoeveel getallen er in een list zitten.

Zoals beloofd is er ook nog een __continue__ instructie. Waar je __break__ gebruikt om een loop in zijn geheel af te breken, gebruik je __continue__ om de code voortijdig te stoppen en door te laten gaan naar een volgende ronde door de loop. Kijk maar eens naar het volgende voorbeeld:

```python
# We maken een lijstje met getallen.

lijstje = [0, 1, 0, 0, 2, 0, 0, 400, 0, 0, 20]

# We zijn geïnteresseerd in 400 gedeeld door n,
# waarbij n een getal uit ons lijstje is.
# Maar, zoals je weet is delen door 0
# niet toegestaan in de wiskunde!

for item in lijstje:
    if item == 0:
        # Niet delen door nul!
        continue
    print(400 / item)
    
    # En alles op dit niveau wordt ook
    # overgeslagen door continue.

```

Zodra __item__ 0 is, gaat de __for__ loop snel door naar het volgende item. Op deze manier delen we nooit door nul. Uiteraard kun je het bovenstaande ook helemaal regelen met __if__ and __else__, maar __continue__ is een elegant mechanisme om grote blokken code over te slaan.

## Belangrijke tip

Een stuk code is niet sneller omdat jij denkt dat je het super efficient hebt getypt. Een stuk code is alleen sneller dan een ander stuk code als je hebt gemeten dat het consistent sneller is.

```python
# Een voorbeeld van onleesbare code 
# die wel werkt.

import random
print(random.SystemRandom().choice(range(10,100,5)))
```

Toegegeven, het heeft een zekere 'cool'-factor om dit zo te programmeren, vooral ook omdat het werkt, maar uiteindelijk gaat het er om of je programma werkt en of jij of iemand anders er over vijf jaar nog steeds soep van kan koken. Dus onderdruk de neiging zo te programmeren, want het is een nachtmerrie om fouten te vinden in programmacode die zo is gemaakt. Daarom hieronder hoe je precies dezelfde code kunt schrijven maar dan leesbaar.

```python
# Gebruik random module om willekeurig item
# uit lijst te selecteren.

import random

# random.SystemRandom() is nog willekeuriger
# dan reguliere random() functie.
random_selector = random.SystemRandom()

# Maak lijst van getallen tussen 10 en 100
# in stappen van 5. 
getallen_lijst = range(10, 100, 5)

# Selecteer een willekeurig getal.
n = random_selector.choice(getallen_lijst)

# Geef weer.
print(n)
```

En stel dat we duizend keer zo'n willekeurig getal willen selecteren... welke code is dan sneller? Als je duizend keer iets moet selecteren in een loop, ja, dan is het misschien wel interessant om te weten welke code sneller uitgevoerd wordt. Als je bijvoorbeeld een computerspel programmeert, waarbij de achtergrond continu getekend moet worden, dan wordt de gameplay mogelijk gehinderd door code die niet snel genoeg is. Meten is weten, dus laten we de bovenste twee voorbeelden eens meten, okay?

```python
# Eerst zetten we beide code fragmenten
# in hun eigen functie definitie.

import random


def code1():
    """ Overdreven efficiente versie van code """
    print(random.SystemRandom().choice(range(10,100,5)))

    
def code2():
    """ Super leesbare versie van code """
    # random.SystemRandom() is nog willekeuriger
    # dan reguliere random() functie.
    random_selector = random.SystemRandom()

    # Maak lijst van getallen tussen 10 en 100
    # in stappen van 5. 
    getallen_lijst = range(10, 100, 5)

    # Selecteer een willekeurig getal.
    n = random_selector.choice(getallen_lijst)

    # Geef weer.
    print(n)
```

Dit bovenstaande stuk Python doet nog geen jota, maar nu hebben we tenminste iets wat we een heleboel keer in een loop kunnen aanroepen en timen.

```python
import random

def code1():
    """ Overdreven efficiente versie van code """
    print(random.SystemRandom().choice(range(10,100,5)))

# iPython manier om code te meten.
%timeit code1()

# Output op mijn computer:
# 10000 loops, best of 3: 116 µs per loop
```

```python
import random

def code2():
    """ Super leesbare versie van code """
    # random.SystemRandom() is nog willekeuriger
    # dan reguliere random() functie.
    random_selector = random.SystemRandom()

    # Maak lijst van getallen tussen 10 en 100
    # in stappen van 5. 
    getallen_lijst = range(10, 100, 5)

    # Selecteer een willekeurig getal.
    n = random_selector.choice(getallen_lijst)

    # Geef weer.
    print(n)
    
# iPython manier om code te meten.
%timeit code2()

# Output op mijn computer:
# 10000 loops, best of 3: 115 µs per loop
```

Dat is grappig! De code is zo'n beetje even snel. En dat is ook best wel logisch, want als jij geen variabele maakt om je antwoord in op te slaan, dan maakt Python nog altijd wel iets dergelijks voor jou in de achtergrond.

__%timeit__ werkt alleen in iPython en Jupyter notebooks, en is een manier om de snelheid van een functie te meten.

Als je de bovenstaande code een paar keer uitvoert, zul je zien dat er niet altijd hetzelfde getal komt. Zo zou je tot de conclusie kunnen komen dat code1 of code2 misschien toch sneller is. Maar dat is allemaal niet belangrijk. Als dit echte kritieke programmacode was, dan zou je misschien concluderen dat het printen naar het scherm onnodig is omdat het teveel tijd kost:
